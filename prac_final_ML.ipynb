# Introducción al problema

En este trabajo usamos una base de *Kaggle*, con datos de *Spotify* segregados por década. Los datos son distintas variables/medidas de canciones obtenidas por el autor a través de la *API Web* de *Spotify* y la variable objetivo es *target*, que toma valores 0 (*flop*) o 1 (*hit*). Así, nuestro objetivo es resolver un problema de clasificación sobre si las canciones van a ser éxitos (*hits*) o no (*flops*), que podría ser usado por *Spotify* para introducir o no canciones en sus listas de éxitos y recomendaciones a sus usuarios, mejorando la experiencia de uso de estos. Un posible KPI podría ser el porcentaje de canciones que van a ser un éxito de una *playlist*.

Las variables del dataset son:
- **track**: nombre de la pista.


- **artist**: nombre del artista.


- **uri**: identificador de la pista.


- **danceability**: describe qué tan adecuada es una pista para bailar. Un valor de `0.0` es menos bailable `1.0` es más bailable.


- **energy**: una medida de `0.0` a `1.0` y representa una medida perceptiva de intensidad y actividad.


- **key**: la clave general estimada de la pista. Los números enteros se asignan a tonos utilizando la notación de clase de tono estándar. P.ej. `0 = C `, `1 = C?/D?`, `2 = D `, y así sucesivamente. Si no se detectó ninguna clave, el valor es `-1`.


- **loudness**: el volumen general de una pista en decibelios (dB). Los valores típicos oscilan entre `-60` y`0` db. 


- **mode**: indica la modalidad (mayor o menor) de una pista, el tipo de escala de la que se deriva su contenido melódico. Mayor está representado por `1` y menor es `0`.


- **speechiness**: detecta la presencia de palabras habladas en una pista. Cuanto más parecida a la voz sea la grabación más cerca de `1.0` será el valor del atributo. 


- **acousticness**: una medida de confianza de `0.0` a `1.0` de si la pista es acústica.


- **instrumentalness**: Predice si una pista no contiene voces. Los sonidos “Ooh” y “aah” se tratan como instrumentales en este contexto. Cuanto más cerca esté el valor de instrumentalidad de `1.0`, mayor será la probabilidad de que la pista no contenga contenido vocal.


- **liveness**: detecta la presencia de una audiencia en la grabación. Los valores de vivacidad más altos representan una mayor probabilidad de que la pista se interprete en vivo. 


- **valence**: una medida de `0.0` a `1.0` que describe la positividad musical transmitida por una pista.


- **tempo**: el tempo general estimado de una pista en pulsaciones por minuto (BPM).  


- **duration_ms**:  la duración de la pista en milisegundos.


- **time_signature**: una firma de tiempo general estimada de una pista. El compás (medidor) es una convención de notación para especificar cuántos tiempos hay en cada compás (o compás).


- **chorus_hit**: esta es la mejor estimación del autor de cuándo comenzaría el coro para la pista. 


- **sections**: el número de secciones que tiene la pista en particular. 


- **target**: la variable de destino para la pista. Puede ser `0` o `1`. `1` implica que esta canción ha aparecido en la lista semanal (emitida por Billboards) de pistas Hot-100 en esa década al menos una vez y, por lo tanto, es un `éxito`. `0` implica que la pista es un `fracaso`.


La condición del autor de que una pista sea fracaso es la siguiente:
 - El tema no debe aparecer en la lista de `hit` de esa década.
 - El artista del track no debe aparecer en la lista de éxitos de esa década.
 - La pista debe pertenecer a un género que podría considerarse no convencional y/o vanguardista.
 - El género de la pista no debe tener una canción en la lista de 'hit'.
 - La pista debe tener EE. UU. como uno de sus mercados.

# Algoritmos de clasificación

### Métricas

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, ConfusionMatrixDisplay, f1_score, RocCurveDisplay, roc_auc_score
```

Las métricas que vamos a usar para validar nuestros modelos son:
- Accuracy

-                                       Precision

-                                       Recall

-                                       Confusion Matrix

-                                       F-SCORE

-                                       Curva ROC

-                                       Área bajo la curva.
<br>
Definimos una función que nos permita evaluar todas a la vez.

```python
def metricas(y_val,y_pred):
  print("Accuracy: %.4f" % accuracy_score(y_val, y_pred))
  print("Precision: %.4f" % precision_score(y_val, y_pred))
  print("Recall: %.4f" % recall_score(y_val, y_pred))
  print("F-score: %.4f" % f1_score(y_val, y_pred))
  print("AUC: %.4f" % roc_auc_score(y_val, y_pred))
  ConfusionMatrixDisplay.from_predictions(y_val, y_pred)
  RocCurveDisplay.from_predictions(y_val, y_pred)
```

Dividimos nuestro conjunto de datos en tres conjuntos:
- Entrenamiento (train)
- Validación
- Test

```python
from sklearn.model_selection import train_test_split
```

```python
def splitting(df):
    X = df.loc[:, df.columns!='target'].values
    y = df['target'].values
    X_t, X_test, y_t, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
    X_train, X_val, y_train, y_val = train_test_split(X_t, y_t, test_size=0.2, random_state=0)
    return X_test, y_test, X_train, y_train, X_val, y_val, X_t, y_t
```

```python
X_test, y_test, X_train, y_train, X_val, y_val, X_t, y_t = splitting(data)
```

### LDA

```python
lda = LDA(n_components=1)
# Usamos una sola componente en LDA debido a que tenemos solo 2 clases (clasificación binaria)
```

```python
lda.fit(X_train, y_train)
y_pred = lda.predict(X_val)
metricas(y_val, y_pred)
```

### kNN

Vamos a probar con distinto número de vecinos para ver cuál es el óptimo dado estos datos

```python
ks = list(range(1, 25))
accs = {}
```

```python
for k in ks:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_val)
    acc = accuracy_score(y_val, y_pred)
    accs[k] = acc
```

```python
ks_arr = np.fromiter(accs.keys(), dtype=int)
accs_arr = np.fromiter(accs.values(), dtype=float)

plt.plot(ks_arr, accs_arr, 'ok')
plt.grid(True)
plt.xlabel('Hiperparámetro $k$');
plt.ylabel('Accuracy sobre validación');
```

Verificamos si obtenemos mejores resultados escalando los datos:

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc1 = StandardScaler()
X_train_sc = sc1.fit_transform(X_train)
X_val_sc = sc1.transform(X_val)
```

```python
for k in ks:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_sc, y_train)
    y_pred = knn.predict(X_val_sc)
    acc = accuracy_score(y_val, y_pred)
    accs[k] = acc
```

```python
ks_arr = np.fromiter(accs.keys(), dtype=int)
accs_arr = np.fromiter(accs.values(), dtype=float)

plt.plot(ks_arr, accs_arr, 'ok')
plt.grid(True)
plt.xlabel('Hiperparámetro $k$');
plt.ylabel('Accuracy sobre validación');
```

Vemos que el máximo en esta rejilla de valores se alcanza a los 22 vecinos, por lo que nos quedamos con ese parámetro.  

**Nota**: Nos parece curioso que se obtengan mejores resultados con *k* par (respecto de sus impares adyacentes) ya que cuando el número de vecinos es par es posible que haya empates y en este caso, la predicción se sortea aleatoriamente. En otros algoritmos como *Random Forest* también hay mucho componente aleatorio y aún así funciona bien, por lo que no nos preocupa demasiado.

```python
knn = KNeighborsClassifier(n_neighbors=17)
knn.fit(X_train_sc, y_train)
y_pred = knn.predict(X_val_sc)
metricas(y_val, y_pred)
```

Para nuestros datos, que no son excesivos en cantidad, hemos podido elegir 22 vecinos, que sin embargo parece un gran número. Si tuviéramos una mayor cantidad de observaciones quizás nos plantearíamos elegir un *k* menor en un *trade-off* entre precisión y coste computacional.

### Neural Networks

Para la red neuronal usamos todos los datos directamente ya que el método *fit* incorpora directamente un argumento para validar.

```python
X = data.loc[:, data.columns!='target'].values
y = data['target'].values
```

```python
# Escalamos los datos para que la red pueda trabajar mejor con ellos
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_sc = sc.fit_transform(X)
```

```python
# Vamos a crear una red con 3 capas ocultas con número de neuronas decreciente hacia la final
# Usamos ReLU como función de activación en las capas ocultas y la función
# sigmoide en la capa final (clasificación binaria)
from tensorflow.keras.models import Sequential
model = Sequential() 

from tensorflow.keras.layers import Dense
number_inputs = 16
number_classes = 2

model.add(Dense(units=256,activation='relu', input_dim=number_inputs))
model.add(Dense(units=128,activation='relu'))
model.add(Dense(units=64,activation='relu'))
model.add(Dense(1, activation='sigmoid')) 

# Compilamos el modelo
# Usamos la entropía cruzada binaria como función de pérdida y el accuracy como métrica
import tensorflow as tf
opt = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])

# Introducimos un criterio de parada
patience = 50
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='accuracy',  patience=patience, restore_best_weights=True)

# Entrenamos la red
history = model.fit(X_sc, y, validation_split=0.2, epochs=50, batch_size=10, shuffle=True, callbacks=[early_stopping], verbose=2)
```

Podemos ver un accuracy en los datos de entrenamiento y de validación de 0.93 y 0.77, respectivamente. Pese a no ser mal resultado, otros algoritmos mucho más simples consiguen resultados similares (como LDA), por lo que igual habría que probar distintos modelos de redes neuronales (más o menos capas, más o menos neuronas, otras funciones de activación/pérdida, etc.) para ver si podemos conseguir una mejora significativa en el rendimiento o si simplemente estos datos funcionan mejor con unos algoritmos que con otros.

**Nota**: Hemos decidido usar 50 *epochs*, ya que habíamos probado con 100 y hemos visto que a pesar de que el *accuracy* de entrenamiento llega hasta el 0.97, el *accuracy* de validación va descendiendo a partir de las 15 *epochs* más o menos, por lo que estábamos cayendo en un sobreajuste de los datos (buen ajuste en entrenamiento pero mala generalización). Quizás reduciendo aún más el número de *epochs* y tocando otros parámetros pudiéramos llegar a un mejor rendimiento de la red.
